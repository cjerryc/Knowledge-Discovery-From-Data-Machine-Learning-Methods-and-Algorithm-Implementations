```python
from math import dist

A = [1, 1]
B = [2, 1]
C = [0, 1]
D = [4, 1]
E = [4, 2]
F = [3, 1]


def calculate_dist(p1, p2):
    return round(dist(p1, p2), 2)


# Find the smallest distance in matrix to cluster on and return the 2 cluster indices
# In the event of ties, the lowest index value takes priority ex. A < F, so A wins the tie
def find_smallest_distance(distMat):
    smallestVal = distMat[len(distMat)-1][len(distMat[len(distMat)-1])-2]       # Sets bottom right non-zero val as temporary smallestVal
    row = len(distMat)-1
    col = len(distMat[row])-2

    for i in reversed(range(len(distMat))):
        for j in reversed(range(len(distMat[i])-1)):
            if distMat[i][j] <= smallestVal:
                if j <= col:
                    smallestVal = distMat[i][j]
                    row = i
                    col = j

    return row, col


def print_matrix(mat, cluster):
    for i in range(len(mat)):
        print(f"{cluster[i]} {mat[i]}")


def single_link_Hierarchical_Clustering():
    point_arr = [A, B, C, D, E, F]
    cluster_arr = [("A",), ("B",), ("C",), ("D",), ("E",), ("F",)]
    distMatrix = []

    for i in range(len(point_arr)):
        distMatrix.append([])
        for j in range(0, i+1):
            distMatrix[i].append(calculate_dist(point_arr[i], point_arr[j]))

    # Print first initial Distance Matrix
    print(f"    {cluster_arr}")
    print_matrix(distMatrix, cluster_arr)

    while len(cluster_arr) > 2:
        # Concatenate row and col tuples into one cluster (merge into "col" cluster), delete residual "row" cluster
        row, col = find_smallest_distance(distMatrix)
        cluster_arr[col] = cluster_arr[col] + cluster_arr[row]
        del cluster_arr[row]
        print("\n")
        print(f"    {cluster_arr}")

        # Find the least of new cluster pair
        temp = [0.0]
        for i in range(len(distMatrix)):
            if i == row or i == col:  # skip the "row" and "col" valued rows in which the rows are to be deleted and don't calculate distances for next iteration.
                continue
            if len(distMatrix[i]) > col:
                if len(distMatrix[i]) > row:
                    if distMatrix[i][col] < distMatrix[i][row]:
                        temp.append(distMatrix[i][col])
                    else:
                        temp.append(distMatrix[i][row])
                else:
                    if distMatrix[i][col] < distMatrix[row][i]:
                        temp.append(distMatrix[i][col])
                    else:
                        temp.append(distMatrix[row][i])
            else:
                if len(distMatrix[i]) > row:
                    if distMatrix[col][i] < distMatrix[i][row]:
                        temp.append(distMatrix[col][i])
                    else:
                        temp.append(distMatrix[i][row])
                else:
                    if distMatrix[col][i] < distMatrix[row][i]:
                        temp.append(distMatrix[col][i])
                    else:
                        temp.append(distMatrix[row][i])

        # Delete the row of distance matrix of row, create new merged column based on shortest distances, then delete old columns
        # Del columns
        for i in range(len(distMatrix)):
            if row > col:
                if len(distMatrix[i]) > row:
                    del distMatrix[i][row]
                if len(distMatrix[i]) > col:
                    del distMatrix[i][col]
            if col > row:
                if len(distMatrix[i]) > col:
                    del distMatrix[i][col]
                if len(distMatrix[i]) > row:
                    del distMatrix[i][row]
        # Del rows
        del distMatrix[row]
        del distMatrix[col]

        distMatrix.insert(0, [temp[0]])
        for i in range(1, len(temp)):
            distMatrix[i].insert(0, temp[i])

        print_matrix(distMatrix, cluster_arr)




single_link_Hierarchical_Clustering()

```
