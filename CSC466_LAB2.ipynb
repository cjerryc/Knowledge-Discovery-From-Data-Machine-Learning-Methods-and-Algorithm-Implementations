```python
from math import dist, fsum
from statistics import mean

k = 2
starting_centers = [(1, 5), (2, 5)]


def k_means_clustering(data_list):
    centroids = starting_centers
    while True:
        cen_assign = [[], []]
        # assign data points to centroids for 1 iteration
        for i in range(len(data_list)):
            dist1 = dist(centroids[0], data_list[i])
            dist2 = dist(centroids[1], data_list[i])
            if dist1 <= dist2:
                cen_assign[0].append(data_list[i])
            else:
                cen_assign[1].append(data_list[i])
        # recalculate centroids as mean of points
        # print(centroids)
        c1, c2 = calculate_centroids(cen_assign)
        if c1 == centroids[0] and c2 == centroids[1]:
            return centroids, cen_assign
        else:
            centroids[0] = c1
            centroids[1] = c2
    # cen_assign = [[], []]
    # # assign data points to centroids for 1 iteration
    # for i in range(len(data_list)):
    #     dist1 = dist(centroids[0], data_list[i])
    #     dist2 = dist(centroids[1], data_list[i])
    #     if dist1 <= dist2:
    #         cen_assign[0].append(data_list[i])
    #     else:
    #         cen_assign[1].append(data_list[i])
    # # recalculate centroids as mean of points
    # c1, c2 = calculate_centroids(cen_assign)
    # if c1 == centroids[0] and c2 == centroids[1]:
    #     return centroids, cen_assign
    # else:
    #     centroids[0] = c1
    #     centroids[1] = c2


# data_assign is a 2_d Vector, as 0th index belongs to centroid1 and 1st index belongs to centroid 2
def calculate_centroids(data_assign):
    # calculate new centroids, given the datapoint assignments
    c1_new_x = []
    c1_new_y = []
    c2_new_x = []
    c2_new_y = []
    # print(data_assign)
    for x in data_assign[0]:
        c1_new_x.append(x[0])
        c1_new_y.append(x[1])
    for y in data_assign[1]:
        c2_new_x.append(y[0])
        c2_new_y.append(y[1])
    return (mean(c1_new_x), mean(c1_new_y)), (mean(c2_new_x), mean(c2_new_y))


def run_k_means():
    data_coord = [(1, 1), (2, 1), (0, 1), (4, 1), (4, 2), (3, 1)]
    centers, assignments = k_means_clustering(data_coord)
    print(f"Final Centroids and Datapoint Assignments at \n {centers[0]}: {assignments[0]} \n {centers[1]}: {assignments[1]}")
    return


run_k_means()
```
