```python

def take_input():
    print("Let's Calculate the probability it will be a good day to play tennis!")
    Outlook = ["sunny", "overcast", "rain"]
    Temperature = ["hot", "mild", "cool"]
    Humidity = ["high", "normal"]
    Wind = ["weak", "strong"]

    print("Outlook: ")
    outlook = str(input()).lower()
    if outlook not in Outlook:
        print("error")
        return
    print("Temperature: ")
    temperature = str(input()).lower()
    if temperature not in Temperature:
        print("error")
        return
    print("Humidity: ")
    humidity = str(input()).lower()
    if humidity not in Humidity:
        print("error")
        return
    print("Wind: ")
    wind = str(input()).lower()
    if wind not in Wind:
        print("error")
        return
    print(f"The day is characterized by Outlook: {outlook}, Temperature: {temperature}, Humidity: {humidity}, Wind: {wind}.")
    return outlook, temperature, humidity, wind


def naive_bayes(outlook, temperature, humidity, wind):
    # P(Play | Outlook, Temperature, Humidity, Wind)
    p_playtennis = 9/ 14
    p_noplaytennis = 5 / 14

    p_condition_play = {"sunny": 2 / 9, "overcast": 4 / 9, "rain": 3 / 9, "hot": 2 / 9, "mild": 4 / 9, "cool": 3 / 9, "high": 3 / 9, "normal": 6 / 9, "weak": 6 / 9, "strong": 3 / 9}
    p_condition_noplay = {"sunny": 3 / 5, "overcast": 0, "rain": 2 / 5, "hot": 2 / 5, "mild": 2 / 5, "cool": 1 / 5, "high": 4 / 5, "normal": 1 / 5, "weak": 2 / 5, "strong": 3 / 5}

    num = p_condition_play[outlook] * p_condition_play[temperature] * p_condition_play[humidity] * p_condition_play[wind] * p_playtennis
    den1 = num
    den2 = p_condition_noplay[outlook] * p_condition_noplay[temperature] * p_condition_noplay[humidity] * p_condition_noplay[wind] * p_noplaytennis

    return num / (den1 + den2)


def probability_program():
    outlook, temperature, humidity, wind = take_input()
    probability = naive_bayes(outlook, temperature, humidity, wind)
    print(f"The probability we play tennis Today is: {round(probability, 3)}")
    return


probability_program()
```
